```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train , y_test = train_test_split(X,Y, test_size = 0.2 , random_state = 40)
```

```python
display(X_train.shape)
display(y_train.shape)
```

```python
m_test = len(y_test)
y_test = y_test.reshape(m_test,1)

m_train = len(y_train)
y_train = y_train.reshape(m_train,1)

X_train = np.concatenate((np.ones((m_train,1)), X_train ), axis = 1 )
X_test = np.concatenate((np.ones((m_test,1)), X_test ), axis = 1 )
```

```python
def sigmoid_fun(X , theta):
    sig = 1/(1+np.exp((-1)*np.dot(X,theta)))
    return sig

def cost(theta,X,y,lamb):
    m = len(y)
    n = len(theta)
    y = y.reshape(m,1)
    theta = theta.reshape(n,1)
    
    h = sigmoid_fun(X , theta)
    
    j = -np.dot(y.T,np.log(h)) - (np.dot((1-y).T,(1-np.log(h))))
    
    J =  j/(m)
    J = J + ( lamb*np.sum( (theta[1:,:]**2) ) )/(2*m)
    
    return J
def grad (theta,X,y,lamb):
    n = len(theta)
    m = len(y)
    y = y.reshape(m,1)
    theta = theta.reshape(n,1)
    grad = np.dot(X.T, (sigmoid_fun(X,theta) - y))/m
    grad[1:,:] = grad[1:,:] + (lamb/m)*theta[1:,:]
    return grad
```

```python
import scipy.optimize as op
theta = np.zeros((X_train.shape[1],1) )

lamb = 0

result = op.fmin_tnc(cost, x0=theta, args=(X_train,y_train, lamb),fprime=grad)
finaltheta = result[0].reshape((-1,1))
```

```python
y_pred = sigmoid_fun(X_test , finaltheta)
#y_pred
len(y_pred)
```

```python
for i in range(len(y_pred)):
    if (y_pred[i] >= 0.5):
        y_pred[i] = 1
    else:
        y_pred[i] = 0
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix( y_test , y_pred)
```

```python
#y_test
from sklearn.metrics import classification_report

print(classification_report(y_test , y_pred))
```

### Using KNN

```python
def dist( X1 , X2):
    return (np.sum((X1-X2)**2))**0.5
```

```python
def final_label(dist_label , k):
    count_0 = 0
    float_k = float(k)
    for i in range(k):
        if dist_label[i,1] == 0:
            count_0 = count_0 + 1
    
    if count_0 > float_k/2 :
        return 0
    else:
        return 1
```

```python
def knn( X_train, X_test , y_train, y_test, k = 5):
    
    y_pred_labels = []
    
    m_test = len(y_test)
    m_train = len(y_train)
    
    for i in tqdm(range(m_test)):
        
        dist_label = []
        
        for j in range(m_train):
            d = dist(X_train[j],X_test[i])
            dist_label.append([d, y_train[j]])
            
        dist_label = sorted(dist_label)
        dist_label = dist_label[:k]
        dist_label = np.array(dist_label)
        
        y_pred_labels.append(final_label(dist_label,k))
            
    print(y_pred_labels)
    
    y_pred_labels = np.asarray(y_pred_labels)
    y_pred_labels =np.reshape(y_pred_labels, (-1, 1))
    print(y_pred_labels.shape)
    return y_pred_labels  
    
    
```

```python
y_pred_knn = knn( X_train, X_test , y_train, y_test, k = 5)
```

```python
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix( y_test , y_pred_knn)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test , y_pred_knn))
```
